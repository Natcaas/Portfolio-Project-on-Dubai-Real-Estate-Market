5. **Missing values**

```python
df_copy.isnull().sum()

# gives us understanding how much of missing values are there in the data (Latitude/Longitude have missing data)
```

6.**Delete Unnessary columns**

```python
df_clean_copy=df_copy.drop(['Latitude','Longitude','Furnishing'], axis=1)
```

```python
df_clean_copy.head(1)
```

7. **Add a new Column**

```python
print(df_clean_copy['Posted_date'].dtypes)         # I want to have just a year column/ checking data type first
```

```python
df_clean_copy['Year']=df_clean_copy['Posted_date'].dt.year
```

```python
df_clean_copy.head(2)
```

8.**Sorting**

```python
df_clean_copy.sort_values(by=['Rent','City'], ascending = True).head(19)
# were found 15 zero values (Outliers) in a DF within 'Rent' column
```

9.**Cleaning**

```python
# since there shouldn't be the value as 0,1,290 (Outliers) I will replace them with the mean.
# first I will create a 'mean variable' and then use replace.mean method
df_clean_copy_mean= df_clean_copy['Rent'][(df_clean_copy['Rent']!=0) & (df_clean_copy['Rent']!=1)
                    & (df_clean_copy['Rent']!=290)].mean()
df_clean_copy['Rent']=df_clean_copy['Rent'].replace ([0,1,290], df_clean_copy_mean)
df_clean_copy.sort_values(by='Rent', ascending=True).head(3)
```

9.**Filtering**

```python
## rearrange columns in this order
df_clean_copy =df_clean_copy[['Address', 'Location', 'City', 'Rent', 'Beds','Baths', 'Type','Area_in_sqft',
                              'Rent_per_sqft', 'Rent_category', 'Posted_date', 'Year', 'Age_of_listing_in_days']]
df_clean_copy.head(3)
```

```python
## filter to Dubai and apartments & create a new variable df_dubai.
df_city = df_clean_copy['City']=='Dubai'
df_type = df_clean_copy['Type']=='Apartment'
df_dubai = df_clean_copy[df_city & df_type]
df_dubai.head(3)

## variable df_dubai is ready to work with
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dubai_rent_count=df_dubai.groupby(['Year','Rent']).size().reset_index(name='Count')
sns.lineplot(dubai_rent_count, x='Year',y='Count',hue='Rent', marker=0)
plt.show()

### this bar graph shows that in the past 2 years tourists of Dubai doubled in number mostly in the first half of the year
```

```python
## In spite of booming tourism and rentals in demand ,
#there's still a problem with some of the locations as the listings exceed 3 years.

df_dubai_age=df_dubai[['Location','Rent','Age_of_listing_in_days']]
df_dubai_age.sort_values(by=['Age_of_listing_in_days'], ascending = False)
```
